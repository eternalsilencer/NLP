```python
import requests
from bs4 import BeautifulSoup
import re
import os
```

```python
url = "http://www.bjsubway.com/station/zjgls/"
response = requests.get(url)
response.encoding = "gbk"
html = response.text
soup = BeautifulSoup(html, 'lxml')
```

```python
def get_txt_name():
    txt_src_name = []
    for i in range(5, 10):
        txt_src_name += soup.find_all('td', {'colspan': str(i)})
    return txt_src_name
# print(get_txt_name())
    
                   
```

```python
def get_txtuseful_name():
    obj = []
    for each in get_txt_name():
        obj += re.findall(r">(.+)<", str(each))
    return obj
print(get_txtuseful_name())
```

```python
Stationinfo = soup.find_all('tbody')  # 汤汁是ResultSet，即结果集
def get_stationinfo():
    obj = []
    for each in Stationinfo:
        temp = re.findall(r">(.+?)<", str(each))   # 正则匹配，str格式
        obj += temp
    return obj
```
