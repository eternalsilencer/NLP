```python
import pandas as pd
import jieba
import pickle
from sklearn.feature_extraction.text import  TfidfVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
fname = 'D:/Google Downloads/export_sql_1558435/sqlResult_1558435.csv'
content = pd.read_csv(fname, encoding='gb18030')
```

```python
content = content.dropna()
```

```python
content.head()
```

```python
content['lable'] = content['source'] == '新华社'

content[content['source'] == '新华社']
```

```python
content_list = content.content.values.tolist()
```

```python
len(content_list)
```

```python
content_S = [' '.join(jieba.cut(s)) for s in content_list]
# for line in content_list:
#     current_segment = jieba.lcut(line)
#     if len(current_segment) > 1 and current_segment != ("\r\n"):
#         content_S.append(' '.join(current_segment))
```

```python
with open('content_S.txt','wb') as fw:
    pickle.dump(content_S, fw, -1)
# Pickle the list using the highest protocol available. 
```

```python
with open('content_S.txt','rb') as fr:
    content_S = pickle.load(fr) 
```

```python
len(content_S)
```

```python
#从文件导入停用词表
with open('./stopwords.txt', 'rb') as fp:
    stopword = fp.read().decode('utf-8')  # 提用词提取
#将停用词表转换为list  
stpwrdlst = stopword.splitlines()
```

```python
tv=TfidfVectorizer(stop_words=stpwrdlst, max_features=300)
data = tv.fit_transform(content_S)
```

```python
pd.DataFrame(data.toarray(), columns=tv.get_feature_names()).head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.toarray(), content['lable'], test_size=0.33, random_state=42)
```

```python
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X_train, y_train) 

```

```python
neigh.score(X_test, y_test)
```
